머신러닝으로 선형 함수 학습하기

- 사용자 데이터를 이용해 y=2x-1 학습하기

```python
import tensorflow as tf
import numpy as np
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
```

레이어는 1개, 뉴런은 1개, 입력은 1개인 매우 간단한 신경망

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

**loss**: 실제 값과 예측 값이 얼마나 다른 지 계산

**optimizer**: loss 를 최소화하여 더 나은 예측값을 내도록 하는 함수

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
model.fit(xs, ys, epochs=500)
```

`fit` 함수를 이용해서 신경망을 학습시킴. `epoch` 만큼 반복하여 학습.

```python
model.predict([10.0])
```

`predict` 함수로 실제 예측을 함. 2x-1 이니까 19 에 근사한 값이 출력됨
